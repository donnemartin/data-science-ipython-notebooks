```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris_dataframe = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                 columns= iris['feature_names'] + ['target'])
```

```python
iris_dataframe['species'] = pd.Categorical.from_codes(iris.target, 
                            iris.target_names)
```
